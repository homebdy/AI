```python
import os
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
```

```python
boston = load_boston()
```

```python
X = boston['data']
Y = boston['target']
```

```python
boston.feature_names
```

```python
data = np.concatenate([X, Y.reshape(-1,1)], axis = 1)
```

```python
data = pd.DataFrame(data)
```

```python
data.head()
```

```python
names = np.append(boston.feature_names, 'Y')
```

```python
data.columns = names
```

```python
data.head()
```

```python
data.shape
```

```python
#data X, Y 가르기
data_X = data.iloc[:, :-1]
data_Y = data.iloc[:, -1]
```

```python
num_train = round(0.7*data.shape[0])
```

```python
num_train
```

```python
# train, test 데이터 나눔
train_X, test_X, train_Y, test_Y = train_test_split(data_X, data_Y, test_size = data.shape[0] - num_train, random_state=42)
```

```python
num_vali = round(0.7*num_train)
```

```python
num_vali
```

```python
train_X.shape
```

```python
test_X.shape
```

```python
# validation 데이터 나눔
train_X, vali_X, train_Y, vali_Y = train_test_split(train_X, train_Y, test_size = num_train - num_vali, random_state=42)
```

```python
train_X.shape
```

```python
vali_X.shape
```

```python
test_X.shape
```

```python
from sklearn.neighbors import KNeighborsRegressor as knn
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_percentage_error as MAPE
```

```python
K = [1, 3, 5, 7, 9, 11]
MAPE_SET = []

for k in K:
    KNN = knn(n_neighbors = k)
    KNN.fit(train_X, train_Y)
    pred = KNN.predict(vali_X)
    MAPE_SET.append(MAPE(vali_Y, pred)*100)
```

```python
MAPE_SET
```

```python
best = np.where(MAPE_SET == min(MAPE_SET))[0][0]
best_k = K[best]
```

```python
best_k
```

```python
best_knn = knn(n_neighbors = best_k)
basic_knn = knn()
```

```python
best_knn.fit(test_X, test_Y)
basic_knn.fit(test_X, test_Y)
predict_best = best_knn.predict(test_X)
predict_basic = basic_knn.predict(test_X)
```

```python
print(MAPE(test_Y, predict_basic)*100)
print(MAPE(test_Y, predict_best)*100)
```

```python
# train data로 테스트
```

```python
train_best_knn = knn(n_neighbors = best_k)
train_basic_knn = knn()
```

```python
train_best_knn.fit(train_X, train_Y)
train_basic_knn.fit(train_X, train_Y)
train_predict_best = best_knn.predict(train_X)
train_predict_basic = basic_knn.predict(train_X)
```

```python
print(MAPE(train_Y, train_predict_basic)*100)
print(MAPE(train_Y, train_predict_best)*100)
```
